```python
import requests
r = requests.get('https://hitomi.la/alltags-a.html')
```

```python
from bs4 import BeautifulSoup as bs
content = bs(r.content, "html.parser").prettify()
print(content)
```

```python
tag_expression = r'\/tag.+.html">\n\s+(.+)\n\s+</a>\n\s+(\(\d+\))'
import re
print(re.findall(tag_expression, content))
```
